```python
#naming the years

years=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(years,yield_apples);
```

```python
#plotting a line graph between years and yield of apples
plt.plot(years,yield_apples)

#naming the x axis
plt.xlabel('Years')

#naming the y axis
plt.ylabel('Yield(tons per hectare)');
```

```python
#representing years in between 2000 and 2002
years=range(2000,2003)
apples=[0.895,0.91,0.919]
oranges=[0.962,0.941,0.930]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)');
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

#naming the title for the entire graph
plt.title('Crop Yields in Kanto')

#naming or representing the lines with names
plt.legend(['Apples','Oranges']);
```

```python
#creates the marker for apples as o that is circle
plt.plot(years,apples,marker='o')

#creates the marker for oranges as x that is cross mark
plt.plot(years,oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
#naming the apple as a solid line and width as 2 points and marker as a circle
plt.plot(years,apples,marker='o',c='b',ls='-',lw=2)

#naming the oranges as a dashed line and width as 5 points and marker as a x
plt.plot(years,oranges,marker='x',c='r',ls='--',lw=5)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
#representing apples line as solid line with squares and in blue color
plt.plot(years,apples,'s-b')

#representing oranges line as dashed line with circles as markers and in red color
plt.plot(years,oranges,'o--r')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
#representing apples as only squares and in blue color without specifying the type of line
plt.plot(years,apples,'sb')

#representing oranges line as only circles and in red color without specifyint the type of line
plt.plot(years,oranges,'or')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
#changing the size of the figure
plt.figure(figsize=(12,6))

#representing oranges with circles and in red color without specifying the type of line
plt.plot(years,oranges,'or')
plt.title('Yield of Oranges(tons per hectare)');
```

```python
# Data Visualization using Python
# Dot Plot

import matplotlib.pyplot as plt

# plotting using plt.pyplot()
plt.plot([4,7,3,6,1,8,9,2,3], 'ro')

# axis labeling
plt.xlabel('numbers')
plt.ylabel('values')

# figure name
plt.title('Dot Plot : Red Dots')

##########################################
##########################################

plt.figure()
# function for a different figure

# plotting using plt.pyplot() Figure 2
plt.plot([1,7,3,6,8,42,34,62],[1,8,9,3,10,11,12,13], 'go')

# axis labeling
plt.xlabel('numbers')
plt.ylabel('values')

# figure name
plt.title('Dot Plot : Green Dot')

##########################################
##########################################

plt.figure()
# function for a different figure

# plotting using plt.pyplot() Figure 2
plt.plot([1,7,3,6,8,42,34,62],[1,8,9,3,10,11,12,13], 'yo')

# axis labeling
plt.xlabel('numbers')
plt.ylabel('values')

# figure name
plt.title('Dot Plot : Yellow Dot')
```
